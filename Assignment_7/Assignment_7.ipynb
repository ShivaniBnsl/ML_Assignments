# Data Field Explanation:

  Explanation of Columns:
.Fixed Acidity:Acid keeps the wine at the fresh crisp.Without Acidity,the wine is flat and thats why fixed Acidity 
dataset is used.
.Volatile Acidity:It is basically a wine spoilage.
.Citric Acid:This dataset revels about the Sourness of wine.
.Residual Sugar:It is usually measured in grams of sugar per litre of wine.
.Chlorides:It increases the concentration.
.Free sulfur dioxide:Amount of Sulfur dioxide.
.total sulfur dioxide:It is the portion of SO2 that is free in the wine plus the portion that is bound to other chemicals in the wine such as aldehydes, pigments, or sugars.
.density:mass per unit volume.
.ph:It is a scale of acidity from 0 to 14. It tells how acidic or alkaline a substance is.
.Sulphates:These are basically the salts.
.alcohol:containing ethanol.
.quality:whether the wine is good or bad in taste.

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)

'''
NOTE:I is used yo finding the 
     distribution of data and then keeping some threshold to 
     identify the outlier.
'''
```

```python
# Identifies outliners 
print(df < (Q1 - 1.5 * IQR))
```

```python


# Remove outliners
df_o = df[(z < 3).all(axis=1)]
df_out = df_o[~((df_o < (Q1 - 1.5 * IQR)) |(df_o > (Q3 + 1.5 * IQR))).any(axis=1)]
df_out.shape
```
