```python
# Question 6
sns.heatmap(df1.isnull(), cbar=True, cmap='gray', yticklabels=False)
plt.title('White space signifies missing data', loc='left', fontweight='bold', y=1.02)
```

```python
# Plot distribution of ages by Survived
sns.distplot(df1['Age'][df1['Survived']==1], hist=False, bins=range(0,80,5), color='gray', label='Survived')
sns.distplot(df1['Age'][df1['Survived']==0], hist=False, bins=range(0,80,5), color='orange', label='Died')
plt.title('Density Plot of Age by Survived', loc='left', fontweight='bold', y=1.02)
```

```python
sns.countplot(x='Survived', hue='Sex', data=df1, palette='RdBu_r')
plt.title('Survival by Sex', loc='left', fontweight='bold', y=1.02)
```
